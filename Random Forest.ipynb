```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import timeit
import pandas
df = pd.read_excel('data.xlsx', index_col=0) 
df.head()
```

```python
#only nummeric features
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
df = df.select_dtypes(include=numerics)
#split data into train en test
unique_game_id = df.game_id.unique()
train_game_id , test_game_id= train_test_split(unique_game_id,test_size=0.30, random_state=42)

test_data = df[df.game_id.isin(test_game_id)]
train_data = df[df.game_id.isin(train_game_id)]
xtest = test_data.drop("bResult",axis=1)
xtest = xtest.drop("rResult",axis=1)
ytest = test_data.bResult
xtrain = train_data.drop("bResult",axis=1)
xtrain = xtrain.drop("rResult",axis=1)
ytrain = train_data.bResult
```

```python
ytest5 = test_data[test_data.minute<6].bResult
xtest5 = test_data[test_data.minute<6]

xtest5 = xtest5.drop("bResult",axis=1)
xtest5 = xtest5.drop("rResult",axis=1)

xtest5 = xtest5.select_dtypes(include=numerics)

ytest10 = test_data[test_data.minute<11][test_data.minute>5].bResult
xtest10 = test_data[test_data.minute<11][test_data.minute>5]

xtest10 = xtest10.drop("bResult",axis=1)
xtest10 = xtest10.drop("rResult",axis=1)

xtest10 = xtest10.select_dtypes(include=numerics)

ytest15 = test_data[test_data.minute<16][test_data.minute>9].bResult
xtest15 = test_data[test_data.minute<16][test_data.minute>9]

xtest15 = xtest15.drop("bResult",axis=1)
xtest15 = xtest15.drop("rResult",axis=1)

xtest15 = xtest15.select_dtypes(include=numerics)

ytest20 = test_data[test_data.minute<21][test_data.minute>15].bResult
xtest20 = test_data[test_data.minute<21][test_data.minute>15]

xtest20 = xtest20.drop("bResult",axis=1)
xtest20 = xtest20.drop("rResult",axis=1)

xtest20 = xtest20.select_dtypes(include=numerics)

ytest25 = test_data[test_data.minute<26][test_data.minute>20].bResult
xtest25 = test_data[test_data.minute<26][test_data.minute>20]

xtest25 = xtest25.drop("bResult",axis=1)
xtest25 = xtest25.drop("rResult",axis=1)

xtest25 = xtest25.select_dtypes(include=numerics)
```

```python
ytrain5 = train_data[train_data.minute<6].bResult
xtrain5 = train_data[train_data.minute<6]

xtrain5 = xtrain5.drop("bResult",axis=1)
xtrain5 = xtrain5.drop("rResult",axis=1)

xtrain5 = xtrain5.select_dtypes(include=numerics)
ytrain10 = train_data[train_data.minute<11][train_data.minute>5].bResult
xtrain10 = train_data[train_data.minute<11][train_data.minute>5]

xtrain10 = xtrain10.drop("bResult",axis=1)
xtrain10 = xtrain10.drop("rResult",axis=1)

xtrain10 = xtrain10.select_dtypes(include=numerics)
ytrain15 = train_data[train_data.minute<16][train_data.minute>10].bResult
xtrain15 = train_data[train_data.minute<16][train_data.minute>10]

xtrain15 = xtrain15.drop("bResult",axis=1)
xtrain15 = xtrain15.drop("rResult",axis=1)

xtrain15 = xtrain15.select_dtypes(include=numerics)
ytrain20 = train_data[train_data.minute<21][train_data.minute>15].bResult
xtrain20 = train_data[train_data.minute<21][train_data.minute>15]

xtrain20 = xtrain20.drop("bResult",axis=1)
xtrain20 = xtrain20.drop("rResult",axis=1)

xtrain20 = xtrain20.select_dtypes(include=numerics)
ytrain25 = train_data[train_data.minute<26][train_data.minute>20].bResult
xtrain25 = train_data[train_data.minute<26][train_data.minute>20]

xtrain25 = xtrain25.drop("bResult",axis=1)
xtrain25 = xtrain25.drop("rResult",axis=1)

xtrain25 = xtrain25.select_dtypes(include=numerics)
```

```python
from sklearn.ensemble import RandomForestClassifier
accuracy5 = []
for j in range(10):
    accuracy5.append([])
    for i in range(5):
        clf = RandomForestClassifier(max_depth=j+1, random_state=i)
        clf.fit(xtrain, ytrain)
        accuracy5[j].append( clf.score(xtest5,ytest5) )
    
print(accuracy5)
```

```python
import matplotlib.pyplot as plt

plt.figure()
#plt.plot(accuracy25)
plt.plot([sum(x)/5 for x in accuracy5][0:8])
plt.show()
```

```python

plt.figure()
#plt.plot(accuracy25)
plt.plot([x+1  for x in range(8)],[sum(x)/5 for x in accuracy10][0:8])

plt.show()
```

```python
accuracy5 = []
for j in range(15):
    accuracy5.append([])
    for i in range(5):
        clf = RandomForestClassifier(max_depth=j+1, random_state=i)
        clf.fit(xtrain, ytrain)
        accuracy5[j].append( clf.score(xtest5,ytest5) )

plt.figure()
plt.plot([x+1  for x in range(15)],[sum(x)/5 for x in accuracy5])
plt.ylabel("Accuracy")
plt.xlabel("Maximum Depth")
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.set_title("Accuracy/Max_Depth for time interval 1-5")
color = 'tab:red'
ax1.set_xlabel('Maximum Depth')
ax1.set_ylabel('Accuracy', color=color)
ax1.plot([x+1  for x in range(15)],[sum(x)/5 for x in accuracy5], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Standard Deviation', color=color)  # we already handled the x-label with ax1
ax2.plot([x+1  for x in range(15)],[stdev(x) for x in accuracy5], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
