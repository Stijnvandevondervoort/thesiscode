```python
start = timeit.default_timer()
bKills = []
rKills = []

bTowers = []
bInhibs = []
bDragons = []
bBarons = []
bHeralds= []

rTowers = []
rInhibs= []
rDragons= []
rBarons= []
rHeralds= []


for j in range(len(LeagueofLegends)):
    temp_bTowers = list(map(int, sorted(re.findall(r'\d+',LeagueofLegends.bTowers[j])[0::2])))
    temp_bInhibs = list(map(int, sorted(re.findall(r'\d+',LeagueofLegends.bInhibs[j])[0::2])))
    temp_bDragons = list(map(int, sorted(re.findall(r'\d+',LeagueofLegends.bInhibs[j])[0::2])))
    temp_bBarons =list(map(int, sorted(re.findall(r'\d+',LeagueofLegends.bInhibs[j])[0::2])))
    temp_bHeralds= list(map(int, sorted(re.findall(r'\d+',LeagueofLegends.bInhibs[j])[0::2])))
    temp_rTowers = list(map(int, sorted(re.findall(r'\d+',LeagueofLegends.bInhibs[j])[0::2])))
    temp_rInhibs=list(map(int, sorted(re.findall(r'\d+',LeagueofLegends.bInhibs[j])[0::2])))
    temp_rDragons= list(map(int, sorted(re.findall(r'\d+',LeagueofLegends.bInhibs[j])[0::2])))
    temp_rBarons= list(map(int, sorted(re.findall(r'\d+',LeagueofLegends.bInhibs[j])[0::2])))
    temp_rHeralds= list(map(int, sorted(re.findall(r'\d+',LeagueofLegends.bInhibs[j])[0::2])))
    
    for i in range(int(LeagueofLegends.gamelength[j])):
        bKills.extend([sum(i > x for x in [int(x[0]) for x in literal_eval(LeagueofLegends.bKills[i])])])
        rKills.extend([sum(i > x for x in [int(x[0]) for x in literal_eval(LeagueofLegends.rKills[i])])])
        
        bTowers.extend([sum(i > x for x in temp_bTowers)])
        bInhibs.extend([sum(i > x for x in temp_bInhibs)])
        bDragons.extend([sum(i > x for x in temp_bDragons)])
        bBarons.extend([sum(i > x for x in temp_bBarons)])
        bHeralds.extend([sum(i > x for x in temp_bHeralds)])
        rTowers.extend([sum(i > x for x in temp_rTowers)])
        rInhibs.extend([sum(i > x for x in temp_rInhibs)])
        rDragons.extend([sum(i > x for x in temp_rDragons)])
        rBarons.extend([sum(i > x for x in temp_rBarons)])
        rHeralds.extend([sum(i > x for x in temp_rHeralds)])
elapsed = timeit.default_timer() - start
print(elapsed, " seconds")
```
